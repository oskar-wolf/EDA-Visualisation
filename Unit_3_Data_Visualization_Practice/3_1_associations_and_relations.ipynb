```python
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
penguins = sns.load_dataset('penguins')
```

```python
fig, (ax1,ax2,ax3, ax4) = plt.subplots(1,4,figsize=(30,10))

sns.scatterplot(data = penguins,x='bill_length_mm',y='flipper_length_mm', ax = ax1)

sns.scatterplot(data = penguins,x='bill_length_mm',y='flipper_length_mm', ax = ax2, hue ='species')

sns.scatterplot(data = penguins,x='bill_length_mm',y='flipper_length_mm', ax = ax3, size='bill_depth_mm')

sns.scatterplot(data = penguins,x='bill_length_mm',y='flipper_length_mm', ax = ax4, size='bill_depth_mm', hue='species')

plt.show()
```

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
```

```python
x,y = np.random.multivariate_normal(mean,cov,1000).T
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,6))

sns.kdeplot(x=x,y=y, ax=ax1)

ax2.hist2d(x,y,bins=25,cmap='Blues')

ax3.hexbin(x,y,gridsize=25,cmap='Blues')

plt.show()
```

```python
flights = sns.load_dataset('flights')

jan_flights = flights.query("month == 'Jan'")

fig , (ax1,ax2) = plt.subplots(1,2,figsize = (20,6))

sns.lineplot(data=jan_flights,x="year",y='passengers', ax=ax1)

sns.lineplot(data=flights, x='year', y='passengers', ax=ax2)

plt.show()
```
