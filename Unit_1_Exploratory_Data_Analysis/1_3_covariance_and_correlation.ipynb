```python
import matplotlib.pyplot as plt # import matplotlib   
import seaborn as sns # import seaborn  
import pandas as pd 

# %% load sample data
penguins = sns.load_dataset("penguins") # load dataset    
from sklearn.preprocessing import StandardScaler
penguins = penguins.dropna()

numerical_cols = penguins.select_dtypes(include=['float64','int64']).columns
categorical_cols = penguins.select_dtypes(include=['object']).columns

penguins = pd.get_dummies(penguins)

scaler = StandardScaler()
penguins[penguins.columns] = scaler.fit_transform(penguins[penguins.columns])

print(penguins.head())

# %% create a correlogram
plt.figure() # create figure  
sns.set_theme(style="whitegrid")  
sns.heatmap(penguins.corr()) # generate heatmap containing correlation coefficients  
plt.show() # show figure 
```
